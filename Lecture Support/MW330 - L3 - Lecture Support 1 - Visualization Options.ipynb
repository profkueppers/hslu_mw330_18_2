```python
# dataset https://archive.ics.uci.edu/ml/datasets/auto+mpg

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

df = sns.load_dataset("mpg")
```

```python
# Slide 60 - histogram
df.weight.hist(bins=10)
plt.show()
df.weight.hist(bins=40)
plt.show()
```

```python
df.weight.plot.kde()
```

```python
# Slide 61 - histogram
sns.distplot(df.weight, kde=False, rug=True, bins=10)
plt.show()
#85/df.weight.count()
sns.distplot(df.weight, kde=False, rug=True, bins=40)
plt.show()
```

```python
# Slide 61 - histogram
sns.distplot(df.weight, kde=True, rug=True, bins=10)
plt.show()
sns.distplot(df.weight, kde=True, rug=True, bins=40)
plt.show()
```

```python
# Slide 61 - density plot
sns.kdeplot(df.weight,bw=300, alpha=1)
sns.kdeplot(df.weight,bw=50, alpha=0.6)
sns.kdeplot(df.weight,bw=10, alpha=0.2)
plt.show()

```
