```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

# EDA: Citibike Trips

```python
trips = pd.read_parquet("201707-citibike-tripdata.parquet")
trips.head()
```

```python
sns.scatterplot(data=trips, x='start station longitude', y='start station latitude')
```

# DBSCAN (I)

```python
X = trips[['start station latitude', 'start station longitude']]
model = DBSCAN(eps=0.001, min_samples=3, metric="euclidean").fit(X)
labels = model.labels_
```

```python
sns.scatterplot(x=X['start station longitude'], y=X['start station latitude'], hue=labels)
```
